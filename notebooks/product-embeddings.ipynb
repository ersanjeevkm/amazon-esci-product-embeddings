# Preprocessing

```python
import pandas as pd
df_products = pd.read_parquet('shopping_queries_dataset_products.parquet')
```

```python
df_products.columns.tolist()
```

```python
df_products.head()
```

```python
df_products['product_locale'].unique()
```

```python
df_products = df_products[df_products['product_locale'] == 'us']
df_products = df_products.drop('product_locale', axis=1)
```

```python
df_products.head()
```

```python
if df_products['product_id'].nunique() == len(df_products):
    print("All product IDs are unique.")
else:
    print("Product IDs are not unique.")
```

```python
df_products_context = pd.DataFrame()
df_products_context['product_id'] = df_products['product_id']
df_products_context['product_info'] = (
    "Product Title:" + df_products['product_title'].astype(str) +
    "\nProduct Description:" + df_products['product_description'].astype(str) +
    "\nProduct Bullet Points:" + df_products['product_bullet_point'].astype(str) +
    "\nProduct Brand:" + df_products['product_brand'].astype(str) +
    "\nProduct Color:" + df_products['product_color'].astype(str)
)
```

```python
df_products_context.head()
```

```python
df_products_context.iloc[0]['product_info']
```

```python
df_products.to_parquet('products_details.parquet')

df_products_context.to_parquet('products_context.parquet')
```
